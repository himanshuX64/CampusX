# Standardization

```python
import pandas as pd 
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('dataset/housing.csv')
df
```

```python
df.shape
```

```python
x=df.iloc[:,0:2]
y=df.iloc[:,-1]
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(x,y,test_size=0.1,random_state=0)
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
y_train.ndim
```

```python
y_test.shape
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
# fit the scaoler to the train set,it will learn the parameter
scaler.fit(X_train)

#  transform train and test sets
X_train_scaled=scaler.transform(X_train)
X_test_scaled=scaler.transform(X_test)

# y_train_scaled=scaler.transform(y_train)
# y_test_scaled=scaler.transform(y_test)     # Beacuse y_train clumn dimention 1 D so not scale
```

```python
scaler.mean_
```

```python
X_train
```

```python
X_train_scaled
```

```python
X_train_scaled=pd.DataFrame(X_train_scaled,columns=X_train.columns)
X_test_scaled=pd.DataFrame(X_test_scaled,columns=X_test.columns)
```

# why scaling is Important

```python
from sklearn.linear_model import LogisticRegression
```

```python
lr=LogisticRegression()
lr_scaled=LogisticRegression()
```

```python
lr.fit(X_train,y_train)
lr_scaled.fit(X_train_scaled,y_train)
```

```python
y_pred=lr.predict(X_test)
y_pred_scaled=lr_scaled.predict(X_test_scaled)
```

```python
from sklearn.metrics import accuracy_score
```

```python
print("Actual data Accuracy :",accuracy_score(y_test,y_pred))
print("Scale data Accuracy :",accuracy_score(y_test,y_pred_scaled))
```
