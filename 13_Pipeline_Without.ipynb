```python
import pandas as pd
```

```python
df=pd.read_csv('dataset/titanic.csv')
df.head()
```

# Data Exploration

```python
data=df.drop(['PassengerId','Name','Ticket','Cabin'],axis=1)
data
```

```python
data.isnull().sum()
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(data.iloc[:,1:],data.iloc[:,0:1],test_size=0.2,random_state=42)
X_train.head(6)
```

```python
X_train.head()
```

```python
X_train.isnull().sum()
```

```python
y_train.sample(5)
```

# Missing value Fill by SimpleImputer

```python
from sklearn.impute import SimpleImputer
```

```python
si_age=SimpleImputer()  # Age
si_embarked=SimpleImputer(strategy='most_frequent')  # Embarked

X_train_age=si_age.fit_transform(X_train[['Age']])
X_train_embarked=si_embarked.fit_transform(X_train[['Embarked']])

X_test_age=si_age.transform(X_test[['Age']])
X_test_embarked=si_embarked.transform(X_test[['Embarked']])
```

```python
X_train_age
```

```python
X_train_embarked
```

```python
X_test_embarked
```

# OneHot Encoding Sex and Embarked

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
ohe_sex=OneHotEncoder()
ohe_embarked=OneHotEncoder()

X_train_sex=ohe_sex.fit_transform(X_train[['Sex']])
X_train_embarked=ohe_embarked.fit_transform(X_train_embarked)

X_test_sex=ohe_sex.transform(X_test[['Sex']])
X_test_embarked=ohe_embarked.transform(X_test_embarked)
```

```python
X_train_sex=X_train_sex.toarray().astype(int)
X_train_sex
```

```python
X_test_sex=X_test_sex.toarray().astype(int)
```

```python
X_train_embarked=X_train_embarked.toarray().astype(int)
X_train_embarked
```

```python
X_test_embarked=X_test_embarked.toarray().astype(int)
```

# concatenate All Columns

```python
X_train.head(3)
```

```python
X_train_rem=X_train.drop(columns=['Sex','Age','Embarked'])
```

```python
X_test_rem=X_test.drop(columns=['Sex','Age','Embarked'])
```

```python
import numpy as np
```

```python
# print(type(X_train_age), getattr(X_train_age, 'shape', None))
# print(type(X_train_sex), getattr(X_train_sex, 'shape', None))
# print(type(X_train_embarked), getattr(X_train_embarked, 'shape', None))
```

```python
X_train_conc=np.concatenate((X_train_rem,X_train_age,X_train_sex,X_train_embarked),axis=1)
```

```python
X_test_conc=np.concatenate((X_test_rem,X_test_age,X_test_sex,X_test_embarked),axis=1)
```

```python
X_train_conc
```

```python
X_train_conc.shape
```

```python
X_test_conc.shape
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
dtc=DecisionTreeClassifier()
dtc.fit(X_train_conc,y_train)
```

```python
y_pred=dtc.predict(X_test_conc)
```

```python
y_pred
```

```python
from sklearn.metrics import accuracy_score
```

```python
print(accuracy_score(y_test,y_pred))
```
