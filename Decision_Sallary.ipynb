```python
import pandas as pd
data=pd.read_csv("salaries_dt.csv")
data
```

```python
data['company'].replace({'google':1,'abc pharma':2,'facebook':3},inplace=True)
```

```python
data
```

```python
data['job'].replace({'sales executive':1,'business manager':2,'computer programmer':3,'sales executive':4},inplace=True)
```

```python
data
```

```python
data['degree'].replace({'bachelors':1,'masters':2},inplace=True)
```

```python
data
```

```python
X=data.iloc[:,0:3].values
y=data.iloc[:,-1].values
```

```python
X.shape
```

```python
y.shape
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2)
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
model=DecisionTreeClassifier()
```

```python
model.fit(X_train,y_train)
```

```python
y_pred=model.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred)
```

```python
from sklearn.tree import plot_tree
from matplotlib.pylab import rcParams
rcParams['figure.figsize']=80,50
plot_tree(model)
```
