```python
import pandas as pd
df=pd.read_csv("iris.csv")
df
```

```python
df['variety'].unique()
```

```python
df['variety'].replace({'Setosa':'1', 'Versicolor':'2', 'Virginica':'3'},inplace=True)
```

```python
df
```

```python
from sklearn.model_selection import train_test_split
X_train,y_train,X_test,y_test=train_test_split(df[['sepal.length','sepal.width','petal.length','petal.width']],df['variety'],test_size=0.2,random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression
model=LogisticRegression()
model.fit(X_train,y_train)
```

```python
model.predict(X_test)
```
