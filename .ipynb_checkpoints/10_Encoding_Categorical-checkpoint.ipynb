```python
import pandas as pd
```

```python
df=pd.read_csv('dataset/customer.csv')
df.head()
```

```python
df=df.iloc[:,2:]
df.head()
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(df.iloc[:,0:2],df.iloc[:,-1],test_size=0.1,random_state=0)
```

```python
X_train
```

```python
y_train
```

# OrdinalEncoder

```python
from sklearn.preprocessing import OrdinalEncoder
```

```python
oe=OrdinalEncoder(categories=[['M','F'],['Undergrad','Grad','PhD']])
```

```python
oe.fit(X_train)
```

```python
X_train=oe.transform(X_train)
```

```python
X_train
```

```python
oe.categories_
```

# LabelEncoder 

```python
# This Transform should be used encode target value y, and not the Input X
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le=LabelEncoder()
```

```python
le.fit(y_train)
```

```python
le.classes_
```

```python
y_train=le.transform(y_train)
y_test=le.transform(y_test)
```
