```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
air=pd.read_csv('dataset/airlines_flights_data.csv')
air.head(4)
```

```python
tips=sns.load_dataset("tips")
tips.head()
```

```python
flight=sns.load_dataset('flights')
flight
```

```python
iris=sns.load_dataset('iris')
iris
```

 # 1. ScatterPlot (Numerical -Numerical data)

```python
sns.scatterplot(x='total_bill',y='tip',data=tips,hue='sex',style='smoker',size='size') 
```

```python
#  seaborn is keyword value access and in multiple column perferm operation in single  function is called multiple variate
```

# 2. Bar plot(Numerical -categorigal)

```python
sns.barplot(x='class',y='duration',data=air,hue='destination_city')
```

# 3. Box Plot(Numerical- categorical)

```python
sns.boxplot(x='passengers',y='year',data=flight,hue='month')
```

```python
titanic=pd.read_csv('titanic.csv')
titanic.head(2)
```

# 4. Distplot (Numerical -catgorical)

```python
sns.distplot(titanic[titanic['Survived']==0]['Age'],hist=False) # 
sns.distplot(titanic[titanic['Survived']==1]['Age'],hist=False) # show Survive in live 
```

# 5. Heatmap (CAtegorical-categorical)

```python
titanic.head()
```

```python
pd.crosstab(titanic['Pclass'],titanic['Survived'])
```

```python
sns.heatmap(pd.crosstab(titanic['Pclass'],titanic['Survived']))
```

# ClusterMap (Categorical-categorical data)

```python
sns.clustermap(pd.crosstab(titanic['Pclass'],titanic['Survived']))
```
